We've receieved a dataset, the Ames Housing Dataset, containing variables that were used by the Ames, Iowa, Assessor's Office to assess residential properties sold between 2006 to 2010. ### Choosing variables for our model

Since there are nearly 80 variables to consider for our model, we can identify those with stronger positive or negative correlations with our variable of interest, SalesPrice.

```python
train.corr()[['SalePrice']].sort_values(by='SalePrice', ascending = False)[:11]
```

```python
# Creating a correlation heatmap
train_sp_corr = train.corr()[['SalePrice']].sort_values(by='SalePrice', ascending = False)
```

```python
plt.figure(figsize = (2,10))
sns.heatmap(train_sp_corr,
           annot=True,
           cmap='coolwarm',
           vmin=-1)
plt.title('Correlations with Sale Price')
plt.savefig('images/corr_heatmap.jpg', bbox_inches='tight');
```

Overall Qual looks really promising as a predictor of price since its correlation is 0.800. This variable takes into account the overall material and finish of the home. Reviewing the plots above, I will try using Condition 2 and Kitchen Qual in my model since their plots seem to have less overlap in the subcategories. However, these variables are strings so I will need to make dummy variables for them. Neither of these variables have missing values. 

I'm going to check the test dataset to see if all of the condition 2s and kitchen Qual categories are being used. This will help guide which dummy columns will be created/kept.

```python
test['Condition 2'].value_counts()
```

```python
test['Kitchen Qual'].value_counts()
```

Since test has only 3 categories in Condition 2 (Norm, Feedr, PosA) and train has all 8, only the 3 dummy columns contained in test will be kept. For Kitchen Qual, test has all 5 categories in its colum while train has 4. I'll drop another column in train. 

```python
train = pd.get_dummies(train, columns=['Condition 2', 'Kitchen Qual'], drop_first = True)
```

```python
# Checking that the dummy columns were added
train.head()
```

```python
# Removing the other condition 2 dummy columns from the dataset
train.drop(columns=['Condition 2_PosN', 'Condition 2_RRAe', 'Condition 2_RRAn', 'Condition 2_RRNn'], inplace=True)
```

```python
# Checking that the specified dummy columns were removed
train.head()
```

Now that the data have been cleaned, we can save it as a new CSV and begin modeling!

```python
train.to_csv('Datasets/train_clean.csv', index=False)
```
